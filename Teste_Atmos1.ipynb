## Importando bibliotecas necessárias

```python
import pandas as pd
from datetime import datetime
```

## Carregando dados

```python
df_postos = pd.read_csv('tbl_postos.csv')
df_vazao = pd.read_csv('tbl_vazao.csv')
```

## Unindo os dataframes de forma que as colunas num_posto de ambos se correspondam

```python
result = pd.merge(df_postos,df_vazao,how='left',on='num_posto')
#result
```

## Removendo linhas não relevantes e add uma coluna com o cálculo de ENA

```python
#localizando e excluindo linhas com valores iguais a zero na coluna (bin_ena)
result_remove = result.loc[(result['bin_ena']==0)]
ultimo_df = result.drop(result_remove.index)
#criando uma nova coluna com o calculo das ENAS
ultimo_df['ENAS'] = ultimo_df['val_vaz_natr']*ultimo_df['val_produt']
#convertendo a coluna dat_medicao em datetime
ultimo_df['dat_medicao'] =  pd.to_datetime(ultimo_df['dat_medicao'])
#ultimo_df
```

## Add uma coluna apenas com os meses

```python
ultimo_df['mes'] = ultimo_df['dat_medicao'].dt.month
#ultimo_df
```

## Fazendo a média mensal por subsistema

```python
ultimo_df.groupby(['mes', 'nom_ssis'])['nom_ssis','ENAS'].mean()
```
