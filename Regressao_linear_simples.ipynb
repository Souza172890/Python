```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from   sklearn.linear_model import LinearRegression
from   sklearn.metrics import r2_score
import statsmodels.api as sm
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
reduzido = pd.read_excel('/content/drive/MyDrive/PSU_2021/TC550TN_MODFAQ_B1.xlsx')
```

```python
print(reduzido.shape)
```

```python
reduzido.head()
```

```python
print(reduzido.describe())
```

```python
X = reduzido['tempo'].values.reshape(-1,1)
y = reduzido['umidade'].values.reshape(-1,1)


reg = LinearRegression()
reg.fit(X, y)
print("O modelo é: modelo= {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
f_previsaoes = reg.predict(X)


plt.figure(figsize = (16,8))
plt.scatter(
    reduzido['tempo'], 
    reduzido['umidade'], 
    c='blue')


plt.plot(
    reduzido['tempo'],
    f_previsaoes,
    c='black',
    linewidth=3,
    linestyle=':'
)

plt.xlabel(" Tempo ")
plt.ylabel(" Umidade")
plt.show()
```
