## Importando biblioteca

```python
import pandas as pd
```

## Carregando dados

```python
df = pd.read_excel('ACOMPH_12.05.2020.xlsx')
#df.head(5)
```

## Clonando dataframe original

```python
df_clone = df[:]
```

## Filtrando dados referente ao posto de interesse para criação de uma novo dataframe

```python
#filtrando dados referente ao posto de interesse para criação de uma novo dataframe
df1 = df_clone.loc[4:33,['Unnamed: 0','Posto',1]]
df1['Unnamed: 0'] =  pd.to_datetime(df1['Unnamed: 0'])
df1['Nom_Posto'] = df['Unnamed: 1'][0]
df1['Num_Posto'] = df.columns[8]
df1 = df1.rename(columns={'Unnamed: 0': 'Data','Posto':'Vazão_Incremental_Consolidada',1:'Vazão_Natural_Consolidada'})
#df1
```

## Criando o segundo dataframe

```python
df2 = df_clone.loc[4:33,['Unnamed: 0','Posto.12',211]]
df2['Unnamed: 0'] =  pd.to_datetime(df2['Unnamed: 0'])
df2['Nom_Posto'] = df['Unnamed: 9'][0]
df2['Num_Posto'] = df.columns[16]
df2 = df2.rename(columns={'Unnamed: 0': 'Data','Posto.12':'Vazão_Incremental_Consolidada',211:'Vazão_Natural_Consolidada'})
#df2
```

## Concatenando os 2 dataframes

```python
result = pd.concat([df1, df2], axis=0)
result
```
